#Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent them?

Hint: Discuss bias-variance tradeoff, cross-validation, and regularization techniques.

Answer: In Machine Learning, the main goal of a model is to learn useful patterns from the training data and perform well on unseen (new) data. When the model becomes too complex or too simple, its performance drops. These two situations are called overfitting and underfitting.

a. Overfitting

Overfitting happens when a model learns too much from the training data, including noise, random fluctuations, and unnecessary details. As a result, the model performs extremely well on training data but poorly on new data.

Characteristics of Overfitting

Very high training accuracy

Low validation/test accuracy

Model memorizes patterns instead of generalizing

Causes of Overfitting

Model is too complex (deep neural networks, deep decision trees)

Too many parameters

Not enough training data

Training for too many epochs

Lack of regularization

How to Detect Overfitting

Large gap between training and validation accuracy (Training ↑, Validation ↓)

Validation loss starts increasing while training loss keeps decreasing

Learning curves diverge

How to Prevent Overfitting

Cross-Validation (e.g., k-fold CV) Helps evaluate model performance on multiple data splits.

Regularization Techniques

L1 (Lasso)

L2 (Ridge)

Dropout (for neural networks)

Early Stopping Stop training when validation loss starts rising.

Reduce Model Complexity Use smaller trees, fewer layers, or lower polynomial degree.

Data Augmentation Especially in image/text data.

Increase Training Data More data helps the model generalize better.

2. Underfitting

Underfitting happens when the model is too simple to learn the underlying patterns in the data. It performs poorly on both training and test data.

Characteristics of Underfitting

Low training accuracy

Low validation accuracy

Model fails to learn important patterns

Causes of Underfitting

Model is too simple (e.g., shallow decision tree, low-degree polynomial)

Too much regularization

Not enough training time

Important features missing

How to Detect Underfitting

Training accuracy is low

Validation accuracy is also low

Learning curves show both losses are high

How to Prevent Underfitting

Increase Model Complexity Deeper trees, more layers, higher polynomial degree.

Reduce Regularization Lower L1/L2 penalty.

Train Longer More epochs or more iterations.

Add Better Features Feature engineering or using more relevant variables.

3. Bias–Variance Tradeoff

Underfitting = High Bias + Low Variance (Model too simple)

Overfitting = Low Bias + High Variance (Model too complex)

A good model balances both bias and variance to achieve optimal performance.

#Question 3:How would you handle missing values in a dataset? Explain at least three methods with examples.

Hint: Consider deletion, mean/median imputation, and predictive modeling.

Answer:Handling missing values is a key step in data preprocessing because missing data can reduce model accuracy and lead to biased predictions. There are multiple methods to handle missing values depending on the data type and the amount of missing data.

1.Deletion Methods

a) Listwise Deletion (Removing Rows)

Rows containing missing values are removed.

Example: If only 3 out of 1000 rows have missing values, deleting them will not affect the dataset much.

When to use:

Missing data is very small (<5%)

Dataset is large

b) Column Deletion

Entire columns with too many missing values can be removed.

Example: If a column has 70% missing values and is not important → drop the column.

2.Imputation Methods

a) Mean / Median / Mode Imputation

Replace missing values with statistical measures.

Mean → for numerical data (e.g., age)

Median → for skewed numerical data

Mode → for categorical data

Example: If the “Age” column has missing values, replace them with the mean age of all available entries.

b) Forward Fill / Backward Fill

Used mostly in time-series data.

Forward Fill: Fill missing value with the previous value

Backward Fill: Fill missing value with the next value

Example: Stock prices dataset often uses forward fill to maintain continuity.

3.Advanced Techniques

a) KNN Imputation

Uses the values of similar (nearest) data points to fill the missing values.

Example: If height is missing, KNN finds the nearest neighbors with similar age/weight and estimates height.

b) MICE (Multiple Imputation by Chained Equations)

Creates multiple models to predict missing values using other features.

Example: Missing “salary” can be predicted using “age”, “education”, “experience”, etc.



#Question 4:What is an imbalanced dataset? Describe two techniques to handle it
#(theoretical + practical).
Hint: Discuss SMOTE, Random Under/Oversampling, and class weights in models.
In Machine Learning, imbalanced data refers to a dataset where the number of samples in each class is not equal. One class has many more samples than the other(s). This causes the ML model to become biased toward the majority class, because it learns more from that class.

Example of Imbalanced Data

Class Number of Samples

Class 0 (Majority) 950 Class 1 (Minority) 50

Here, Class 0 dominates the dataset. A normal classifier may predict everything as Class 0 and still show 95% accuracy, but it completely fails to detect minority cases.

This is a common issue in:

Fraud detection

Medical diagnosis

Spam detection

Churn prediction

Why Imbalanced Data is a Problem?

Model becomes biased toward majority class

Poor recall for minority class

Accuracy becomes misleading

Rare but important cases are missed (fraud, disease, etc.)

Techniques to Handle Imbalanced Data

Below are two commonly used and effective techniques:

SMOTE (Synthetic Minority Oversampling Technique)
SMOTE increases the size of the minority class by creating synthetic samples, not just duplicating existing ones.

How SMOTE Works

It selects a minority class sample

Finds its nearest neighbors

Creates new synthetic data points between them

Advantages

Prevents overfitting compared to simple oversampling

Works well with continuous data

Use Case Example

In fraud detection, SMOTE helps to balance fraudulent vs. non-fraudulent transactions.

Random Undersampling
This method reduces the majority class size by randomly removing samples.

How Undersampling Works

Selects majority class

Randomly deletes some data points

Balances both classes

Advantages

Fast and simple

Reduces memory usage

Works well when dataset is large

Disadvantages

May remove important samples

Can reduce model accuracy

Other Techniques.

Class Weighting
Assign higher weight to the minority class during model training. Used in:

Logistic Regression

Random Forest

SVM

Neural Networks

Ensemble Methods
Balanced Random Forest

EasyEnsemble These combine multiple models to reduce bias.

Cost-Sensitive Learning
Penalizes the model more for misclassifying the minority class.




#Question 5: Why is feature scaling important in ML? Compare Min-Max scaling and Standardization.

Hint: Explain impact on distance-based algorithms (e.g., KNN, SVM) and gradient descent.

Answer: Feature scaling means bringing all numerical features into a similar range (like 0–1 or -1 to 1). It helps ML models treat all features equally and improves model accuracy and training speed.

 Impact of Feature Scaling on ML Algorithms

1.Prevents large-value features from dominating
Example: Salary (0–1,000,000) vs Age (18–70)

Without scaling → salary dominates all calculations.

2.Improves distance-based algorithms
Algorithms that depend on distances get more accurate after scaling.

3.Helps gradient-based algorithms converge faster
Gradient descent moves smoothly when all features are on the same scale.

Algorithms that NEED Scaling

Algorithm Reason

KNN Uses distance → scaling mandatory K-Means Cluster formation depends on distances SVM Margin + distance calculations Logistic Regression Uses gradient descent Neural Networks Stable gradients, faster training

Algorithms that DO NOT Need Scaling

Algorithm Reason

Decision Tree Splits based on thresholds, not distances Random Forest Tree-based → scaling not required XGBoost / LightGBM Also tree-based

Example

If a dataset has:

Age: 25

Salary: 50,000

Salary is very large compared to age. Distance-based models will treat salary as more important. Scaling fixes this by making both features comparable.

#Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?

Hint: Consider categorical variables with ordinal vs. nominal relationships.

Label Encoding vs. One-Hot Encoding
1. Label Encoding

What it does:

Maps each category to an integer (e.g., Red → 0, Blue → 1, Green → 2).

Pros:

Simple and compact (no increase in dimensionality).

Works well when categories have a natural order.

Cons:

Imposes an artificial ordinal relationship if none exists.

Distance-based models may incorrectly interpret category 2 as “greater than” category 0.

2. One-Hot Encoding

What it does:

Creates 1 binary column per category (e.g., Red → [1,0,0], Blue → [0,1,0], Green → [0,0,1]).

Pros:

No artificial ordering.

Works well for nominal data (unordered categories).

Safe for most ML models (linear models, tree models, neural nets, etc.).

Cons:

Increases dimensionality (problematic with high-cardinality features).

May lead to sparsity.

When to Prefer Each?
✅ Use Label Encoding when:

The categorical variable is ordinal (has meaningful order).
Examples:

Education level (Primary < High School < Bachelor < Master < PhD)

Size (Small < Medium < Large)

Satisfaction rating (1–5)

The algorithm can interpret integers without assuming distance is meaningful (e.g., tree-based models like Random Forest or XGBoost).
Trees split based on thresholds, so ordering is acceptable.

✅ Use One-Hot Encoding when:

The categorical variable is nominal (unordered categories).
Examples:

City names

Colors

Gender

Product IDs (if small)

Using distance-based or linear models where numeric encoding could mislead:

K-NN

SVM

Linear/Logistic Regression

Neural Networks

You want to avoid the model interpreting categories as numeric magnitudes.

| Property                         | Label Encoding           | One-Hot Encoding                 |
| -------------------------------- | ------------------------ | -------------------------------- |
| Best for                         | **Ordinal** categories   | **Nominal** categories           |
| Output                           | One integer per category | Several binary columns           |
| Introduces order?                | **Yes**                  | No                               |
| Dimensionality                   | Low                      | High (risk with many categories) |
| Good for tree models?            | Yes                      | Yes                              |
| Good for linear/distance models? | Often **no**             | Yes                              |


#Question 7: Google Play Store Dataset
a). Analyze the relationship between app categories and ratings. Which categories have the highest/lowest average ratings, and what could be the possible reasons?

Dataset: https://github.com/MasteriNeuron/datasets.git

Answer:

```python
import pandas as pd
import requests

# Define the URL for the raw CSV file from the GitHub repository
csv_url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/googleplaystore.csv"

# 1. Load the dataset
# Try to download the file if it's not present
try:
    df = pd.read_csv("googleplaystore.csv")
except FileNotFoundError:
    print(f"Downloading googleplaystore.csv from {csv_url}...")
    response = requests.get(csv_url)
    response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)
    with open("googleplaystore.csv", "wb") as f:
        f.write(response.content)
    df = pd.read_csv("googleplaystore.csv")

# 2. Check basic information
print(df.head())
print(df.info())

# 3. Clean rating column
# Convert "Rating" column to numeric
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Remove the row where 'Category' is '1.9' as it's an erroneous entry
df = df[df['Category'] != '1.9']

# Remove rows with missing rating or category
df = df.dropna(subset=['Rating', 'Category'])

# 4. Calculate average rating by Category
category_ratings = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)

# 5. Show result
print("Average Rating by Category:\n")
print(category_ratings)

# 6. Highest & Lowest rated categories
highest = category_ratings.idxmax()
lowest = category_ratings.idxmin()

print("\nHighest Rated Category:", highest)
print("Lowest Rated Category:", lowest)
```

```python
import pandas as pd
import requests
import matplotlib.pyplot as plt

# Define the URL for the raw CSV file for the Titanic dataset
csv_url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/titanic.csv"

# 1. Load the dataset (Titanic dataset)
# Try to download the file if it's not present
try:
    df = pd.read_csv("titanic.csv")
except FileNotFoundError:
    print(f"Downloading titanic.csv from {csv_url}...")
    response = requests.get(csv_url)
    response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)
    with open("titanic.csv", "wb") as f:
        f.write(response.content)
    df = pd.read_csv("titanic.csv")

# 2. Basic information
print(df.head())
print(df.info())

# Question 8a: Compare survival rates based on passenger class (Pclass)
print("\n--- Question 8a: Survival Rates by Pclass ---")
survival_by_pclass = df.groupby('Pclass')['Survived'].mean().sort_values(ascending=False)
print("Survival rate by Pclass:\n", survival_by_pclass)

highest_pclass_survival = survival_by_pclass.idxmax()
print(f"\nHighest survival rate: Pclass {highest_pclass_survival} ({survival_by_pclass.max():.2%})")
print("Possible reason: First-class passengers likely had better access to lifeboats and were prioritized during evacuation due to their social status.")

# Question 8b: Analyze how age affected survival (Children vs. Adults)
print("\n--- Question 8b: Survival Rates by Age Group ---")
# Create age groups
df['AgeGroup'] = df['Age'].apply(lambda x: 'Child' if x < 18 else 'Adult')

survival_by_age_group = df.groupby('AgeGroup')['Survived'].mean().sort_values(ascending=False)
print("Survival rate by Age Group:\n", survival_by_age_group)

child_survival = survival_by_age_group.get('Child', 0)
adult_survival = survival_by_age_group.get('Adult', 0)

if child_survival > adult_survival:
    print(f"\nChildren had a better chance of survival ({child_survival:.2%}) compared to adults ({adult_survival:.2%}).")
    print("Possible reason: The 'women and children first' protocol was likely followed during the evacuation, leading to higher survival rates for children.")
elif adult_survival > child_survival:
    print(f"\nAdults had a better chance of survival ({adult_survival:.2%}) compared to children ({child_survival:.2%}).")
else:
    print("\nChildren and adults had similar chances of survival.")


# The commented out code below was for the flight price dataset and is no longer relevant for Question 8.
# # 3. How price changes with days left until departure
# # price_by_days = df.groupby("days_left")["price"].mean()

# # print("\nAverage price by days left until departure:")
# # print(price_by_days.head())

# # # Plot price trend (optional if you want visualization)
# # plt.figure(figsize=(8,5))
# # plt.plot(price_by_days.index, price_by_days.values)
# # plt.xlabel("Days Left Until Departure")
# # plt.ylabel("Average Price")
# # plt.title("Price Trend vs Days Left")
# # plt.show()


# # 4. Compare airline prices for same route (Example: Delhi to Mumbai)
# # route_df = df[(df["source_city"] == "Delhi") & (df["destination_city"] == "Mumbai")]

# # airline_price = route_df.groupby("airline")["price"].mean().sort_values()

# # print("\nAverage price by Airline (Delhi → Mumbai):")
# # print(airline_price)


# # 5. Cheapest & Most Expensive airline
# # cheapest = airline_price.idxmin()
# # expensive = airline_price.idxmax()

# # print("\nCheapest Airline:", cheapest)
# # print("Most Expensive Airline:", expensive)
```

#Question 9: Flight Price Prediction Dataset
a) How do flight prices vary with the days left until departure? Identify any exponential price surges and recommend the best booking window.

b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are consistently cheaper/premium, and why?

Dataset: https://github.com/MasteriNeuron/datasets.git

```python
import pandas as pd
import requests
import matplotlib.pyplot as plt

# Define the URL for the raw CSV file from the GitHub repository
csv_url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/flight_price.csv"

# 1. Load dataset
try:
    df = pd.read_csv("flight_price.csv")
except FileNotFoundError:
    print(f"Downloading flight_price.csv from {csv_url}...")
    response = requests.get(csv_url)
    response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)
    with open("flight_price.csv", "wb") as f:
        f.write(response.content)
    df = pd.read_csv("flight_price.csv")

# 2. Basic information
print(df.head())
print(df.info())

# 3. How price changes with days left until departure
price_by_days = df.groupby("days_left")["price"].mean()

print("\nAverage price by days left until departure:")
print(price_by_days.head())

# Plot price trend (optional if you want visualization)
plt.figure(figsize=(8,5))
plt.plot(price_by_days.index, price_by_days.values)
plt.xlabel("Days Left Until Departure")
plt.ylabel("Average Price")
plt.title("Price Trend vs Days Left")
plt.show()


# 4. Compare airline prices for same route (Example: Delhi to Mumbai)
route_df = df[(df["source_city"] == "Delhi") & (df["destination_city"] == "Mumbai")]

airline_price = route_df.groupby("airline")["price"].mean().sort_values()

print("\nAverage price by Airline (Delhi \u2192 Mumbai):")
print(airline_price)


# 5. Cheapest & Most Expensive airline
cheapest = airline_price.idxmin()
expensive = airline_price.idxmax()

print("\nCheapest Airline:", cheapest)
print("Most Expensive Airline:", expensive)
```
